# Imports

Para utilizar as bibliotecas , precisamos importá-las. Isso significa, de um modo simples, trazer o seu código para o nosso script. Vamos importar algumas bastante utilizadas para ciência de dados:

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

Notou algo de estranho?

É possível importar somente apenas algumas funções ou até mesmo alguns submódulos que você irá utilizar, como no caso de `matplotlib.pyplot`. Como ~os programadores são preguiçosos~ esse nome é relativamente grande para ficar repetindo e deixa o código largo, podemos vinculá-lo a uma abreviação, para facilitar as coisas, utilizando a _keyword_ `as`, seguido da abreviação, no caso `plt`.

É uma boa prática importar o mínimo necessário para evitar conflitos de nomes.

Pronto! Agora já podemos utilizá-las e executar esse exemplo da documentação oficial de [matplotlib](https://matplotlib.org/3.2.1/tutorials/introductory/sample_plots.html).

```python
# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='About as simple as it gets, folks')
ax.grid()

plt.show()
```
